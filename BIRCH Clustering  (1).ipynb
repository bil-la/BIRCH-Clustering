```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch

# Generate synthetic data with 3 centers
X, y = make_blobs(n_samples=500, centers=3, cluster_std=0.7, random_state=42)

# Plot the raw data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c='gray', s=20, edgecolor='k')
plt.title("Generated Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Apply BIRCH clustering
birch = Birch(threshold=0.5, n_clusters=3)
birch.fit(X)
labels = birch.predict(X)

# Plot the BIRCH clusters
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=20, edgecolor='k')
plt.title("BIRCH Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
